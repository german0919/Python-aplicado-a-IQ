```python
import numpy as np
import sympy as sp
from matplotlib import pyplot as plt
from  sklearn.metrics import r2_score
```

```python
#Considerando los datos experimentales que puede concluir respecto al tipo de 
#crecimiento y cinetica microbiana. Justifique sus respuestas
t=np.asarray([0,1,1.5,2,2.5,3,3.5])
X=np.asarray([0.2,0.47,1,2.1,4.42,9.4,11.7])
S=np.asarray([25,24.41,23.28,20.9,15.8,5.2,0])
```

```python
plt.plot(t,X,"bo-",label="X")
plt.plot(t,S,"ro-",label="S")
plt.legend(loc="best")
plt.grid(True)
plt.show()
```

```python
#Linealizacion de los datos X vs t
plt.plot(t,np.log(X),"bo-")
plt.title("Linealización de los datos")
plt.xlabel("t")
plt.ylabel("ln(X)")
plt.grid(True)
plt.show()
```

```python
#Linealizacion de los datos X vs t
plt.plot(t[1:6],np.log(X)[1:6],"bo--")
plt.title("Linealización de los datos")
plt.xlabel("t")
plt.ylabel("ln(X)")
plt.grid(True)
plt.show()
Regta=np.polyfit(t[1:6],np.log(X)[1:6],1) #Ajuste lineal
modela=np.poly1d(Regta)
print("Los coeficientes del polinomio son :",Regta)
R_2a=r2_score(np.log(X)[1:6],modela(t[1:6]))#Encontramos el R^2 segun los datos 
print("R^2 =",R_2a) 
print("La velocidad específica de crecimiento será :",Regta[0])
```

```python
#Estime el coeficiente de rendimiento biomasa/sustrato
plt.plot(S,X,"ro-")
plt.title("Rendimiento")
plt.xlabel("S(g/L)")
plt.ylabel("X(g/L)")
plt.grid(True)
plt.show()
Regt2=np.polyfit(S,X,1) #Ajuste lineal
model2=np.poly1d(Regt2)
print("Los coeficientes del polinomio son :",Regt2)
R_2=r2_score(X,model2(S))#Encontramos el R^2 segun los datos 
print("R^2 =",R_2) 
print("Por tanto, el coeficiente de rendimiento biomasa/sustrato =",-Regt2[0])
```
