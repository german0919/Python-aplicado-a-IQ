```python
import numpy as np
import sympy as sp
from matplotlib import pyplot as plt
from  sklearn.metrics import r2_score
```

```python
#En un bioreactor batch se tienen los siguietes datos
t=np.asarray([0,1,2,3,4,5,5.5,6,7,8])
X=np.asarray([0.1,0.15,.2,.25,.68,1.8,3.0,5.0,5.3,5.3])
S=np.asarray([10.0,9.9,9.8,9.7,8.8,6.5,4.1,0.2,0,0])
#Estime la longitud de la fase de demora
#Linealizacion de los datos X vs t
plt.plot(t,np.log(X),"bo-")
plt.title("Linealización de los datos")
plt.xlabel("t")
plt.ylabel("ln(X)")
plt.grid(True)
plt.show()
print("Por tanto la longitud de la fase estacionaria seria: 7-8 hr")
```

```python

#Util si me pidieran miu maxima
#Se toma la porcion del grafico que corresponde a la fase exponencial
plt.plot(t[3:8],np.log(X[3:8]),"ro-")
plt.title("Fase de crecimiento exponencial")
plt.xlabel("t")
plt.ylabel("ln(X)")
plt.grid(True)
plt.show()
Regtl=np.polyfit(t[3:8],np.log(X[3:8]),1) #Ajuste lineal
model=np.poly1d(Regtl)
print("Los coeficientes del polinomio son :",Regtl)
R2=r2_score(np.log(X[3:8]),model(t[3:8]))#Encontramos el R^2 segun los datos 
print("R^2 =",R2) 
```

```python
#Estime el coeficiente de rendimiento biomasa/sustrato
plt.plot(S,X,"ro-")
plt.title("Rendimiento")
plt.xlabel("S(g/L)")
plt.ylabel("X(g/L)")
plt.grid(True)
plt.show()
Regt2=np.polyfit(S,X,1) #Ajuste lineal
model2=np.poly1d(Regt2)
print("Los coeficientes del polinomio son :",Regt2)
R_2=r2_score(X,model2(S))#Encontramos el R^2 segun los datos 
print("R^2 =",R_2) 
print("Por tanto, el coeficiente de rendimiento biomasa/sustrato =",-Regt2[0])
```
