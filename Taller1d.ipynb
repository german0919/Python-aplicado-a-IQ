```python
import numpy as np
import sympy as sp
from matplotlib import pyplot as plt
from  sklearn.metrics import r2_score
```

```python
t=np.asarray([0,0.2,0.5,1.0,1.5,2.0,2.5,3.0,3.5,4.0,4.5])
X=np.asarray([0.45,0.52,0.65,0.81,1.22,1.77,2.13,3.55,4.02,3.77,2.20])
#Grafico
plt.plot(t,X,'bo-')
plt.xlabel("t(Días)")
plt.ylabel("X (cell/ml)*10^-6")
plt.grid(True)
plt.show()
k=0
miu_stad=[]
for k in list(range(0,len(X))):
    miu_si=(1/np.average(X[k:k+2]))*(((np.diff(X[k:k+2])))/((np.diff(t[k:k+2]))))
    miu_stad.append(miu_si)
print(miu_stad) 
```

```python
#
plt.plot(t,np.log(X),'ro-')
plt.xlabel("t (Días)")
plt.ylabel("lnX (cell/ml)*10^-6")
plt.grid(True)
plt.show()
```

```python
plt.plot(t[1:3],np.log(X)[1:3],'mo-')
plt.xlabel("t (Días)")
plt.ylabel("lnX (cell/ml)*10^-6")
plt.title("Linealización fase de crecimiento exponencial")
plt.grid(True)
plt.show()
Regta=np.polyfit(t[1:3],np.log(X)[1:3],1) #Ajuste lineal
modela=np.poly1d(Regta)
print("Los coeficientes del polinomio son :",Regta)
R_2a=r2_score(np.log(X)[1:3],modela(t[1:3]))#Encontramos el R^2 segun los datos 
print("R^2 =",R_2a) 
print("La velocidad específica de crecimiento será :",Regta[0])
```
