```python
RES_C1CH=RES_C1CH.drop(columns=['visc_exp', 'visc_GN', 'visc_JA', 'visc_MC', 'visc_AVM', 'visc_WT',
       'visc_EH'])
RES_C1CH.iloc[:,0:8].melt(id_vars='T[K]')
```

```python
# Crear el scatterplot
plt.figure(figsize=(10, 6))
sns.scatterplot(data=RES_C1CH.iloc[:,0:8].melt(id_vars='T[K]'),
                x='T[K]', y='value', hue='variable', style="variable", palette='plasma', 
                size="variable",sizes=(45, 30), legend="full")

plt.axhline(y=0, color='black', linestyle='--', linewidth=1)

plt.legend(title='Modelo')
plt.title("Ciclohexano-DBT (0.1503 m)")
plt.xlabel("T/ K")
plt.ylabel('$\mu_{exp}-\mu_{pred}$ /cP')
plt.savefig("Residuales_CHDBTC1.pdf")
plt.savefig("Residuales_CHDBTC1.svg")
# Mostrar el gráfico
plt.show()
```
