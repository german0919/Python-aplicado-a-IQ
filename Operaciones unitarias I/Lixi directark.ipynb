## Retenci√≥n constante r=k ##

```python

#Programa para calculo de %rendimiento ingresando el numero de etapas con retenci√≥n constante
#empieza a correr desde que se encuentra alpha
alpha=8
r=0.4
E_nmin=50

n=int(input('Ingrese el n√∫mero de etapas: '))
i=0 #Contador
#Creaci√≥n de vectores para almacenar los resultados
#---------------------------------------------------
R_=np.zeros(n+1)
NR=np.zeros(n+1)
XR=np.zeros(n+1) 
E_n=np.zeros(n+1) 
Ye_n=np.zeros(n+1) 
NE_n=np.zeros(n+1) 
XM=np.zeros(n+1) 
NM=np.zeros(n+1) 
XR1=np.zeros(n+1) 
YE1=np.zeros(n+1) 
NR1=np.zeros(n+1) 
b=np.zeros(n+1) 
R1=np.zeros(n+1) 
E=np.zeros(n+1) 
d=np.zeros(n+1) 
S=np.zeros(n+1) 
#------------------------------------------------------
R_[0]=250
NR[0]=3
XR[0]=1
x1=[]
y1=[]
for i in list(range(0,n)):
	E_n[i]=alpha*E_nmin
	Ye_n[i]=0
	NE_n[i]=0
    #Calculo del punto de mezcla
	XM[i]=(R_[i]*XR[i]+E_n[i]*Ye_n[i])/(R_[i]+E_n[i])
	NM[i]=(R_[i]*NR[i]+E_n[i]*NE_n[i])/(R_[i]+E_n[i])
	XR1[i]=XM[i]
	YE1[i]=XR1[i]
	NE1=0 #Funci√≥n de arrastre
	NR1[i]=5/2 #Funci√≥n de retenci√≥n
	#Resolvemos el sistema para hallar R1 y E1
	R_1,E_1=sp.symbols("R_1,E_1")
	e1=sp.Eq(R_[i]*XR[i]+E_n[i]*Ye_n[i]-R_1*XR1[i]-E_1*YE1[i],0)
	e2=sp.Eq(R_[i]*NR[i]-R_1*NR1[i]-E_1*NE1,0)
	r=sp.solve([e1,e2],E_1,R_1) #arroja un diccionario con la solucion
	R_1=r[R_1]
	E_1=r[E_1]
    
	#Cerrar el balance
	b[i]=(R_[i]+E_n[i])-(R_1+E_1)
	if b[i]==0:
		print('Cierra el balance')
	R1[i]=R_1
	E[i]=E_1
	sm=(np.sum(E*YE1))
	d[i]=(R_[0]*XR[0]+E_n[i]*Ye_n[i])
    #Calculo del porcentaje de recuperaci√≥n
	S[i]=((sm)/(d[i])*100)
    
	R_[i+1]=R1[i]
	NR[i+1]=NR1[i]
	XR[i+1]=XR1[i]
    #Organizaci√≥n de datos para graficar
	x=[XR[i],Ye_n[0]]
	x1.append(x)
	y=[NR[i],NE_n[0]]
	y1.append(y)    
X=np.linspace(0,1,100)
un_o=np.ones(len(X))    
Nr=5/2*un_o #En este caso se hizo esto por ser constante
#Grafica de las lineas para cada etapa
plt.plot(x1[0],y1[0])    
plt.plot(x1[1],y1[1])    
plt.plot(x1[2],y1[2])    
plt.plot(x1[3],y1[3])    

plt.plot(X,Nr,'m')
plt.xlabel('X,Y')
plt.grid(True)    
plt.ylabel('N')
plt.title('Etapas m√∫ltiples')
plt.show()
print("El porcentaje de recuperaci√≥n es :")
print(S[0:n]) 
```
